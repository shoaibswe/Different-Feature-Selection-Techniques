```python
import pandas as pd 
import sklearn as sk
```

```python
df = pd.read_csv('student-por.csv',sep=',')
df.head(5)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['total']= df.G1.values+df.G2.values+df.G3.values
```

```python
import numpy as np

conditions=[
            df['total'].between(48, 60),
            df['total'].between(40, 47),
            df['total'].between(32, 39),
            df['total'].between(0, 31),
            ]
val = ['Very Good', 'Satisfactory', 'Good', 'Avg']

df['Grade']= np.select(conditions,val,default ='Avg')
```

```python
x = df.iloc[:,:-5]
y=df.iloc[:,-1:]
print(x)
print(y)
```

```python
cats = [i for i in x.columns if x.dtypes[i]=='object']
cats 
```

```python
from sklearn.preprocessing import LabelEncoder
le =  LabelEncoder()
```

```python
for i in cats:
    x[i] = le.fit_transform(df[i].astype(str))
```

```python
x
```

```python
## I'll use Kbest and score functions
from sklearn.feature_selection import SelectKBest

from sklearn.feature_selection import f_classif
from sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif
```

```python
#creating instances
with_fc = SelectKBest(score_func=f_classif)
with_ch2 =  SelectKBest(score_func=chi2)
with_mic =  SelectKBest(score_func=mutual_info_classif)
```

```python
#fitting instances
with_fc.fit(x,y)
with_ch2.fit(x,y)
with_mic.fit(x,y)
```

```python
#################################
#getting score using ***f_classif
wfc_score = pd.DataFrame(with_fc.scores_, columns=['fc Score'])
#gettinh columns name
feature_col_names = pd.DataFrame(x.columns)
#addting both column name and score in one table
imp_cols_f_ci = pd.concat([feature_col_names,wfc_score],axis=1)
#sorutng in descending
imp_cols_f_ci.sort_values(by='fc Score',ascending=False)
imp_cols_f_ci
```

```python
##lets use chi_square
ch2_score = pd.DataFrame(with_ch2.scores_, columns=['ch2 Score'])
feature_col_names = pd.DataFrame(x.columns)
imp_cols_ch2 = pd.concat([feature_col_names,ch2_score],axis=1)
imp_cols_ch2.sort_values(by='ch2 Score',ascending=False)
imp_cols_ch2
```

```python
##lets use mutual_info_classif

mic_score = pd.DataFrame(with_mic.scores_, columns=['mic Score'])
feature_col_names = pd.DataFrame(x.columns)
imp_cols_mic = pd.concat([feature_col_names,mic_score],axis=1)
imp_cols_mic.sort_values(by='mic Score',ascending=False)
imp_cols_mic
```

```python
#WOW SEEMS ALMOST ALL RESULTS ARE SAME!
#we can keep featurese which have  highest marks
```

**### lets try easiner way for mutual info classif**

```python
#importing model
from sklearn.feature_selection import mutual_info_classif
#fitting it
mic_imp = mutual_info_classif(x,y)
```

```python
#Lets create a new dataframe with respect to the priority as upper values
imp = pd.DataFrame(mic_imp,index=x.columns)
#lets rename the column and sort on Importance
imp.columns=['cols_to_select']
imp.sort_values(by='cols_to_select',ascending=False)
imp
```

```python
x.to_csv('raw_features.csv')
y.to_csv('label.csv')
```
